```python
import cv2
import numpy as np
```

```python
def contrast(image):
    
    #Grab the Image Dimensions
    h = image.shape[0]
    w = image.shape[1]
    
    #Define Maximum Factor 
    fac = 6
    
    #Define Factor Step Alongside the Width Values (Left: Maximum Factor, Right: 1)
    facStep = (fac - 1) / w
    
    # loop over the Image, Pixel by Pixel
    for y in range(0, h):
        # Reset the Factor for Each Row from the Begining of that row
        newFac = fac
        
        for x in range(0, w):
            # For each Row, Multiply the Pixel Values with the Factor, and Saturate to 255 if value exceeds 255
            image[y, x][0] = image[y, x][0] * newFac if image[y, x][0] * newFac <= 255 else 255
            image[y, x][1] = image[y, x][1] * newFac if image[y, x][1] * newFac <= 255 else 255
            image[y, x][2] = image[y, x][2] * newFac if image[y, x][2] * newFac <= 255 else 255
            
            # Decrease the Factor by the FactorStepSize alongside the Width
            newFac = newFac - facStep

    return image

###############################################################################################

def task1():
    
    # Read Expected Result Image (Martha Wayne Overlayed by Batman)
    expResult = cv2.imread('Q1O.jpg')
    
    # Read 1st Image (Martha Wayne)
    img1 = cv2.imread('Q11.png')
    rows1,cols1,ch1 = img1.shape
    # Apply the Contrast Function on the 1st Image
    img1 = contrast(img1)
    
    # Read 2nd Image (Batman)
    img2 = cv2.imread('Q12.jpg')
    # Flip the 2nd Image Horizontaly
    img2 = cv2.flip( img2, 1 )
    rows2,cols2,ch2 = img2.shape
    
    # Get 3 Matrix Points for each Image
    pts2 = np.float32([[cols2,0],[0,rows2],[cols2,rows2]])
    pts1 = np.float32([[cols1,0],[0,rows1],[cols1,rows1]])
    
    # Get Affine Matrix Transformation for Image 2 to be Image 1
    M = cv2.getAffineTransform(pts2,pts1)

    # Apply the Transformation Matrix on Image 2 to match Image 1 in Size
    newImg2 = cv2.warpAffine(img2,M,(cols1,rows1))
    rows2,cols2,ch2 = newImg2.shape
    
    # Translate Image 2 by a Factor tx = 200, and ty = 0
    M1 = np.float32([[1,0,200],[0,1,0]])
    newImg2 = cv2.warpAffine(newImg2,M1,(cols1,rows1))
    rows2,cols2,ch2 = newImg2.shape
    
    # Show Original Batman and Transformed Batman Images for 10 Seconds
    cv2.imshow('Original Batman Image', img2)
    cv2.imshow('New Batman Image',newImg2)
    cv2.waitKey(10000)
    cv2.destroyAllWindows()
    
    # Create Image with Weighted Addition of Images 1 and 2, with a Weight Ratio of 0.8:0.2
    dst = cv2.addWeighted(img1,0.8,newImg2,0.2,0)
    
    # Show Expected Output and Actual Result for 10 Seconds
    cv2.imshow('Expected', expResult)
    cv2.imshow('Result',dst)
    cv2.waitKey(10000)
    cv2.destroyAllWindows()
    
###############################################################################################    
    
task1()
```

```python
def choose(image1, image2, minX, maxX, minY, maxY):
    # Loop over all Pixels in the area Defined by the Max Boundries provided in the Parameters for an optimized search
    # rather than searching ireelevant pixels in the entire Image
    for y in range(minY, maxY):
            for x in range(minX, maxX):
                # Check if 1st Picture frame is not black
                if not (image1[y,x][0] == 0 & image1[y,x][1] == 0 & image1[y,x][2] == 0):
                    # If true, place Image 1 Pixel Values in Image 2 at the given X and Y Coordinates
                    image2[y, x][0] = image1[y, x][0]
                    image2[y, x][1] = image1[y, x][1]
                    image2[y, x][2] = image1[y, x][2]
    return image2
```

```python
def task21():
    
    # Read Expected Result Image 1 (Sherlock Picture in Big Room Frame)
    expResult = cv2.imread('Q2O1.jpg')
    
    # Read 1st Image (Sherlock Holmes)
    img1 = cv2.imread('Q21.jpg')
    rows1,cols1,ch1 = img1.shape
    
    # Read 2nd Image (Big Room)
    img2 = cv2.imread('Q22.jpg')
    rows2,cols2,ch2 = img2.shape
   
    # Get 3 Matrix Points for each Image
    pts1 = np.float32([[cols1,0],[0,rows1],[cols1,rows1]])
    
    # Define Max Boundries of Frame in Big Room
    minX = 1218
    maxX = 1308
    minY = 377
    maxY = 515
    
    pts2 = np.float32([[maxX,minY],[minX,maxY],[maxX,maxY]])
    
    # Get Affine Matrix Transformation for Image 1 to be in Image 2
    M = cv2.getAffineTransform(pts1,pts2)

    # Apply the Transformation Matrix on Image 1 to match Image 2 in Size
    newImg1 = cv2.warpAffine(img1,M,(cols2,rows2))
    rows1,cols1,ch1 = newImg1.shape
    
    cv2.imshow('New Image 1',newImg1)
    cv2.imshow('Image 2',img2)
    cv2.waitKey(10000)
    cv2.destroyAllWindows()
    
    # Fill Frame with Sherlock Picture
    res = choose(newImg1, img2, minX, maxX, minY, maxY)
    
    # Show Expected Output and Actual Result for 10 Seconds
    cv2.imshow('Expected', expResult)
    cv2.imshow('Result',res)
    cv2.waitKey(10000)
    cv2.destroyAllWindows()
    
###############################################################################################   
     
task21()
```

```python
def task3():
    
    # Read Expected Result Image 1 (Sherlock Picture in Big Frame)
    expResult = cv2.imread('Q3O2.jpg')

    # Read 1st Image (Sherlock Holmes)
    img1 = cv2.imread('Q21.jpg')
    #img1 = cv2.flip( img1, 1 )
    rows1,cols1,ch1 = img1.shape

    # Read 2nd Image (Big Tilted Frame)
    img2 = cv2.imread('Q31.jpg')
    rows2,cols2,ch2 = img2.shape

    # Get 3 Matrix Points for each Image
    pts1 = np.float32([[cols1,0],[0,rows1],[cols1,rows1], [0,0]])

    # Define Max Boundries of Frame in Image
    minX = 158
    maxX = 468
    minY = 36
    maxY = 390

    pts2 = np.float32([[468,70],[158,390], [464, 355], [162, 36]])

    # Get Perspective Matrix Transformation for Image 1 to be in Image 2
    M = cv2.getPerspectiveTransform(pts1,pts2)

    # Apply the Transformation Matrix on Image 1 to match Image 2 in Size
    newImg1 = cv2.warpPerspective(img1,M,(cols2,rows2))
    rows1,cols1,ch1 = newImg1.shape

    cv2.imshow('New Image 1',newImg1)
    cv2.imshow('Image 2',img2)
    cv2.waitKey(10000)
    cv2.destroyAllWindows()

    # Fill Frame with Sherlock Picture
    res = choose(newImg1, img2, minX, maxX, minY, maxY)

    # Show Expected Output and Actual Result for 10 Seconds
    cv2.imshow('Expected', expResult)
    cv2.imshow('Result',res)
    cv2.waitKey(10000)
    cv2.destroyAllWindows()
 
###############################################################################################  

task3()
```
